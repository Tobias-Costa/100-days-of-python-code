### Challenge

- What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience).

```python
# College major with the highest mid-career salary
max_id = clean_df["Mid-Career Median Salary"].idxmax()
clean_df.loc[max_id]
```

- Which college major has the lowest starting salary and how much do graduates earn after university?

```python
# College major with the lowest starting salary
max_id = clean_df["Starting Median Salary"].idxmin()
clean_df.loc[max_id]
```

- Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?

```python
# College major with the lowest mid-career salary
max_id = clean_df["Mid-Career Median Salary"].idxmin()
clean_df.loc[max_id]
```

### Challenge

- Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile.

```python
highest_potential = clean_df.sort_values("Mid-Career 90th Percentile Salary", ascending=False)
highest_potential.head()
```

- Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.

## Group and pivoting

```python
clean_df.groupby('Group').count()
```

```python
# Specify the numeric columns for mean calculation
numeric_columns = ['Spread', 'Starting Median Salary', 'Mid-Career Median Salary', 'Mid-Career 10th Percentile Salary', 'Mid-Career 90th Percentile Salary']

clean_df.groupby('Group')[numeric_columns].mean()
```
